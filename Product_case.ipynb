### Рассмотрим параметр возраст

```python
print("Минимальный возраст у клиентов:", df_client['age'].min())
print("Максимальный возраст у клиентов:", df_client['age'].max())

print("Минимальный возраст у не-клиентов:", df_no_client['age'].min())
print("Максимальный возраст у не-клиентов:", df_no_client['age'].max())
```

```python
print("Пропуски у клиентов:", df_client['age'].isna().sum())
print("Пропуски у не клиентов:", df_no_client['age'].isna().sum())
```

```python
mean_client = df_client['age'].mean()
median_client = df_client['age'].median()
mode_client = df_client['age'].mode()[0]

mean_no_client = df_no_client['age'].mean()
median_no_client = df_no_client['age'].median()
mode_no_client = df_no_client['age'].mode()[0]

plt.figure(figsize=(20,10))
plt.hist(df_client['age'], alpha=0.5, label='Клиенты', bins=100)
plt.hist(df_no_client['age'], alpha=0.5, label='Неклиенты', bins=100)

plt.axvline(mean_client, color='blue', linestyle='--', label=f'Среднее (Клиенты): {mean_client:.1f}')
plt.axvline(median_client, color='blue', linestyle='-', label=f'Медиана (Клиенты): {median_client:.1f}')
plt.axvline(mode_client, color='blue', linestyle=':', label=f'Мода (Клиенты): {mode_client:.1f}')

plt.axvline(mean_no_client, color='red', linestyle='--', label=f'Среднее (Неклиенты): {mean_no_client:.1f}')
plt.axvline(median_no_client, color='red', linestyle='-', label=f'Медиана (Неклиенты): {median_no_client:.1f}')
plt.axvline(mode_no_client, color='red', linestyle=':', label=f'Мода (Неклиенты): {mode_no_client:.1f}')


plt.title('Возраст клиентов и неклиентов')
plt.xlabel('Возраст')
plt.ylabel('Количество человек')
plt.legend()
plt.savefig('Возраст.png', dpi=300, bbox_inches='tight')
```

Заметна разница в левой части распределения. Довольно большая часть неклиентов - молодые люди 16-24 год, т.е. страшеклассники и студенты

```python
reg15_client = df_client['region'].value_counts().head(15)
reg15_no_client = df_no_client['region'].value_counts().head(15)

regions = reg15_client.index.tolist() + [region for region in reg15_no_client.index if region not in reg15_client.index]

client_counts = [reg15_client.get(region, 0) for region in regions]
no_client_counts = [reg15_no_client.get(region, 0) for region in regions]

fig, ax = plt.subplots(figsize=(12, 7))
x = range(len(regions))

ax.bar(x, client_counts, width=0.4, label='Клиенты')
ax.bar([i + 0.4 for i in x], no_client_counts, width=0.4, label='Не-клиенты')

ax.set_title("Топ-15 регионов среди клиентов и не-клиентов")
ax.set_xlabel("Регион")
ax.set_ylabel("Количество")
ax.set_xticks([i + 0.2 for i in x])
ax.set_xticklabels(regions, rotation=90)
ax.legend()
plt.savefig('Регион.png', dpi=300, bbox_inches='tight')
```

Видно, что в некоторых регионах распределение клиентов и не-клиентов обратное: Краснодарский край, Республика Башкортостан, Ростовская область

*Гипотеза: увелечение рекламы в приведенных выше регионах позволит увеличить количество клиентов Т-Города на 5-10%*
